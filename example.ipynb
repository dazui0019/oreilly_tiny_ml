```python
# Numpy is a math library
import numpy as np
# Pandas is a data manipulation library
import pandas as pd
# Matplotlib is a graphing library
import matplotlib.pyplot as plt
# Math is Python's math library
import math

# Set seed for experiment reproducibility
seed = 1
np.random.seed(seed)
```

```python
x = np.linspace(0, 20, 50)
y_points = 3 * x + 7 + np.random.randn(50) * 5
y_line_1 = 3 * x + 7
y_line_2 = 4 * x + 1
y_line_3 = 3 * x + 2

plt.scatter(x, y_points, label='data points')
plt.plot(x, y_line_1, color='red', label='y = 3x + 7')
plt.plot(x, y_line_2, color='green', label='y = 4x + 1')
# plt.plot(x, y_line_3, color='yellow', label='y_3 = 3x + 2')

for i in range(len(x)):
    plt.plot([x[i], x[i]], [y_points[i], y_line_1[i]], color='gray', linestyle='--', linewidth=1)

# for i in range(len(x)):
#     plt.plot([x[i], x[i]], [y_points[i], y_line_2[i]], color='green', linestyle='--', linewidth=1)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter and Line")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('a')
ax.set_ylabel('k')
ax.set_zlabel('z')
# ax.set_title('z = x^2 + y^2')

plt.show()
```
