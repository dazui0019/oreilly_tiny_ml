# 偏导

```python
# 二元函数
def f(x, y):
    return x**2 + y**2

# 曲面网格
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 选定一点
x0, y0 = 1.0, 1.5
z0 = f(x0, y0)

# 偏导数
df_dx = 2 * x0
df_dy = 2 * y0

fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111, projection='3d')

# 曲面
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.35, edgecolor='none')
ax.plot_wireframe(X, Y, Z, color='gray', linewidth=0.2)

# 标记点
ax.scatter(x0, y0, z0, color='red', s=80, label='Point (a0, k0, z0)')

# 截面曲线1：固定 y = y0，只看 x 变化
x_cross = np.linspace(-3, 3, 200)
y_cross = np.full_like(x_cross, y0)
z_cross_x = f(x_cross, y_cross)
ax.plot(x_cross, y_cross, z_cross_x, color='blue', linewidth=3, label='Section: k = k0')

# 截面曲线2：固定 x = x0，只看 y 变化
y_cross2 = np.linspace(-3, 3, 200)
x_cross2 = np.full_like(y_cross2, x0)
z_cross_y = f(x_cross2, y_cross2)
ax.plot(x_cross2, y_cross2, z_cross_y, color='orange', linewidth=3, label='Section: a = a0')

# x方向切线
t = np.linspace(-1, 1, 50)
x_tan = x0 + t
y_tan = np.full_like(t, y0)
z_tan_x = z0 + df_dx * t
ax.plot(x_tan, y_tan, z_tan_x, color='navy', linestyle='--', linewidth=3, label='Tangent for ∂z/∂a')

# y方向切线
x_tan2 = np.full_like(t, x0)
y_tan2 = y0 + t
z_tan_y = z0 + df_dy * t
ax.plot(x_tan2, y_tan2, z_tan_y, color='darkorange', linestyle='--', linewidth=3, label='Tangent for ∂z/∂k')

# 坐标轴和标题
ax.set_xlabel('a')
ax.set_ylabel('k')
ax.set_zlabel('z')
ax.set_title('Partial Derivatives with Cross-Section Curves')

# 调整视角
ax.view_init(elev=28, azim=-55)

ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0.6

def f(x, y):
    return x**2 + y**2 + 2 * a * x * y

# 偏导
def fx(x, y):
    return 2 * x + 2 * a * y

def fy(x, y):
    return 2 * y + 2 * a * x

# 曲面
x = np.linspace(-3, 3, 150)
y = np.linspace(-3, 3, 150)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 选定一点
x0, y0 = 1.0, 1.2
z0 = f(x0, y0)

fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111, projection='3d')

# 曲面
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.45, edgecolor='none')

# 点
ax.scatter(x0, y0, z0, color='red', s=70)

# 截面曲线 y = y0
x_cross = np.linspace(-3, 3, 300)
y_cross = np.full_like(x_cross, y0)
z_cross_x = f(x_cross, y_cross)
ax.plot(x_cross, y_cross, z_cross_x, color='blue', linewidth=3, label=r'Section $y=y_0$')

# 截面曲线 x = x0
y_cross2 = np.linspace(-3, 3, 300)
x_cross2 = np.full_like(y_cross2, x0)
z_cross_y = f(x_cross2, y_cross2)
ax.plot(x_cross2, y_cross2, z_cross_y, color='orange', linewidth=3, label=r'Section $x=x_0$')

# x方向切线
t = np.linspace(-0.8, 0.8, 100)
x_tan = x0 + t
y_tan = np.full_like(t, y0)
z_tan_x = z0 + fx(x0, y0) * t
ax.plot(x_tan, y_tan, z_tan_x, 'b--', linewidth=3, label=r'Tangent for $\partial f/\partial x$')

# y方向切线
x_tan2 = np.full_like(t, x0)
y_tan2 = y0 + t
z_tan_y = z0 + fy(x0, y0) * t
ax.plot(x_tan2, y_tan2, z_tan_y, color='darkorange', linestyle='--', linewidth=3,
        label=r'Tangent for $\partial f/\partial y$')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(r'$f(x,y)=x^2+y^2+2axy,\ a=0.6$')
ax.view_init(elev=30, azim=35)
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 耦合参数，越接近 1 耦合越强
a = 0.9

def f(x, y):
    return x**2 + y**2 + 2 * a * x * y

# 对这个二次函数，固定一个变量时，另一个变量的最优解可以直接写出来
# 固定 y，最优 x = -a*y
# 固定 x，最优 y = -a*x
def coordinate_descent(x0, y0, steps=8):
    path = [(x0, y0)]
    x, y = x0, y0

    for _ in range(steps):
        x = -a * y
        path.append((x, y))

        y = -a * x
        path.append((x, y))

    return np.array(path)

# 起点
x0, y0 = 2.5, -2.0
path = coordinate_descent(x0, y0, steps=7)

# 画等高线
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(8, 8))
contours = plt.contour(X, Y, Z, levels=25, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)

# 画轨迹
plt.plot(path[:, 0], path[:, 1], 'o-', color='red', linewidth=2, markersize=5, label='Coordinate Descent Path')

# 标起点和终点
plt.scatter(path[0, 0], path[0, 1], color='blue', s=80, label='Start')
plt.scatter(path[-1, 0], path[-1, 1], color='black', s=80, label='End')
plt.scatter(0, 0, color='green', s=100, label='Minimum')

plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Coordinate Descent on $f(x,y)=x^2+y^2+2axy$' + f'   (a={a})')
plt.legend()
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.show()
```
